```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

url = "/content/drive/MyDrive/IRIS.csv"
column_names = ['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth', 'Species']

df = pd.read_csv(url, header=None, names=column_names)

print("Original Dataset:")
print(df.head())

label_encoder = LabelEncoder()

df['Species_LabelEncoded'] = label_encoder.fit_transform(df['Species'])

print("\nDataset after Label Encoding (Species column transformed to numbers):")
print(df[['Species', 'Species_LabelEncoded']].head())
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

url = "/content/drive/MyDrive/IRIS.csv"
column_names = ['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth', 'Species']

# Read the dataset into a pandas dataframe
df = pd.read_csv(url, header=None, names=column_names)

# Display the first few rows of the dataset
print("Original Dataset:")
print(df.head())

# ----------------- One-Hot Encoding using OneHotEncoder -----------------
# Initialize OneHotEncoder (sparse=False returns a dense matrix)
encoder = OneHotEncoder()

# Fit and transform the 'Species' column
encoded_data = encoder.fit_transform(df[['Species']])

# Convert the encoded data into a DataFrame
encoded_df = pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out(['Species']))

# Concatenate the encoded data with the original dataframe (excluding the original 'Species' column)
df_one_hot_encoded = pd.concat([df.drop('Species', axis=1), encoded_df], axis=1)

# Display the result after One-Hot Encoding
print("\nDataset after One-Hot Encoding using OneHotEncoder:")
print(df_one_hot_encoded.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
import seaborn as sns

# Loading datasets

dataset1 = pd.read_csv("/content/sample_data/california_housing_test.csv")
dataset2 = pd.read_csv("/content/sample_data/california_housing_train.csv")

plt.scatter(dataset1['median_income'], dataset1['median_house_value'])
plt.title('Scatter Plot - Income vs House Value')
plt.xlabel('Median Income')
plt.ylabel('Median House Value')
plt.show()


# Box Plot - To detect outliers in 'median_house_value'
plt.boxplot(dataset1['median_house_value'])
plt.title('Box Plot - House Value')
plt.show()

# Wave Graph (Uniform Curve) - Example using sine wave
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y)
plt.title('Wave Graph')
plt.xlabel('X Axis')
plt.ylabel('Y Axis (Sine Wave)')
plt.show()

plt.hist(dataset1['median_income'], bins=20)
plt.title('Histogram - Median Income')
plt.xlabel('Median Income')
plt.ylabel('Frequency')
plt.show()

plt.pie(dataset1['median_income'].value_counts(), labels=dataset1['median_income'].unique(), autopct='%1.1f%%')
plt.title('Pie Chart - Median Income')
plt.show()

sns.histplot(dataset1['median_house_value'], kde=True)
plt.title('Histogram - Median House Value')
plt.xlabel('Median House Value')
plt.ylabel('Frequency')
plt.show()

sns.pairplot(dataset1[['median_income', 'median_house_value']])
plt.suptitle('Pair Plot - Income vs House Value')
plt.show()

sns.heatmap(dataset1.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

sns.scatterplot(x='median_income', y='median_house_value', data=dataset1)
plt.title('Scatter Plot - Income vs House Value')
plt.show()

# Handling Skewness
print("\nSkewness of numerical columns:")
print(dataset1.skew())

# Outlier Detection using IQR (Interquartile Range)
Q1 = dataset1['median_house_value'].quantile(0.25)
Q3 = dataset1['median_house_value'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 2* IQR
upper_bound = Q3 + 2 * IQR

outliers = dataset1[(dataset1["median_house_value"]< lower_bound) | (dataset1["median_house_value"]> upper_bound)]
print("Outliers:")
print(outliers)
```
